## Define
We need to find out the values which are not valid and delete them.

## Code

```python
import re
```

```python
rates = []

extract_rates = lambda x: rates.append(re.findall(r'(\d+(\.\d+)|(\d+))\/(\d+0)', x, flags=0))

we_rate['text'].apply(extract_rates)

numerator = []
dog_count = []

for item in rates:
    
    # for tweets with no rating, but a picture, so a dog
    if len(item) == 0:
        numerator.append('NaN')
        dog_count.append(1)
        
    # for tweetss with one rating and one dog
    elif len(item) == 1 and item[0][-1] == '10':
        numerator.append(float(item[0][0]))
        dog_count.append(1)
   
    # for group ratings
    elif len(item) == 1: 
        avg = float(item[0][0]) / (float(item[0][-1]) / 10) 
        numerator.append(avg)
        dog_count.append(float(item[0][-1]) / 10)
   
    # for tweets with more than one rating
    elif len(item) > 1: 
        total = 0
        list = []
        for i in range(len(item)):
            if item[i][-1] == '10': #one tweet has the phrase '50/50' so I'm coding to exclude it
                list.append(item[i])
        for rate in list:
            total = total + float(rate[0])
        avg = total / len(item)
        numerator.append(avg)
        dog_count.append(len(item))
   
    # in order to catch bugs
    else:
        numerator.append('Error')
        dog_count.append('Error') 
        
we_rate['rating'] = numerator # not need to also add denominator since they are all 10!
we_rate['dog_count'] = dog_count
we_rate['rating'].value_counts()
```

## Code

```python
# Firstly, let's store the first true algorithm with it's level of confidence
Prediction = []
Confidence = []

def prediction_confidence(newdataframe):
    if newdataframe['p1_dog']==True:
        Prediction.append(newdataframe['p1'])
        Confidence.append(newdataframe['p1_conf'])
    elif newdataframe['p2_dog']==True:
        Prediction.append(newdataframe['p2'])
        Confidence.append(newdataframe['p2_conf'])
    elif newdataframe['p3_dog']==True:
        Prediction.append(newdataframe['p3'])
        Confidence.append(newdataframe['p3_conf'])
    else:
        Prediction.append('NaN')
        Confidence.append(0)
        
we_rate.apply(prediction_confidence, 1)
we_rate['Prediction'] = Prediction
we_rate['Confidence'] = Confidence
```
